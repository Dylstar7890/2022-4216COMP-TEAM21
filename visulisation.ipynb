```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

spotifydata = pd.read_excel('SpotifyDataBackup.xlsx')

plt.title('popularity over time')
plt.plot(spotifydata.year, spotifydata.popularity, color = '#42ecf5')

plt.xlabel('year')
plt.ylabel('popularity')

plt.show()
```

```python
plt.hist(spotifydata.tempo, color= '#f5bf42')

plt.title('Tempo in Songs')
plt.ylabel('Number of Songs')
plt.xlabel('Tempo of Song (BPM)')

plt.show()
```

```python
spotifydata = pd.read_excel('SpotifyDataBackup.xlsx')

bins = [-50, -40, -30, -20, -10, 0]

plt.hist(spotifydata.loudness, bins=bins, color= '#42f55d')

plt.title('Loundess of a Song')

plt.ylabel('Number of Songs')
plt.xlabel('Loudness of a Song (dB)')

plt.xticks(bins)

plt.show()
```

```python
plt.hist(spotifydata.energy,color= '#c542f5')

plt.title('Measuring the Energy of a Song (dB)')

plt.ylabel('Number of Songs')
plt.xlabel('Energy measured (dB)')

plt.show()
```
